## purl-fetcher API to get druids belonging to Digital Bookplates Collection

```python
import httpx

params = "per_page=10000"
r = httpx.get("https://purl-fetcher.stanford.edu/collections/druid:nh525xs4538/purls", params=params)
print(r.url)
```

```python
r.json()
```

```python
collection = r.json()['purls']
sample = slice(0, 9)

def get_child_druid(x):
    return x['druid']

children_druids = map(get_child_druid, collection[sample])
sample_druids = list(children_druids)
```

We need to lookup each bookplate and parse the xml for title, image filename, Symphony fund name

```python
for druid in sample_druids:
    lookup_druid = druid.replace("druid:", "")
    r = httpx.get(f"https://purl.stanford.edu/{lookup_druid}.xml")
    print(r.url)
    bookplate_xml = r.text
```
